```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder

# Load the dataset
file_path = "new_transaction_csv.csv"
df = pd.read_csv(file_path)

df.rename(columns={"customer_id":"sender_id", "merchant_id":"receiver_id"}, inplace=True)

# Drop unnecessary columns
df_cleaned = df.drop(columns=['transaction_id', 'account_id',  'ip_address', 'timestamp', 'previous_transactions_timestamp'])

# List of categorical columns to encode
categorical_columns = ['merchant_category', 'transaction_type', 'location', 'channel', 'sender_id', 'receiver_id', 'customer_occupation', 'account_type']

# Apply Label Encoding to categorical columns
label_encoders = {}
for col in categorical_columns:
    le = LabelEncoder()
    df_cleaned[col] = le.fit_transform(df_cleaned[col])
    label_encoders[col] = le

# Save the transformed dataset
df_cleaned.to_csv("transformed_transactions.csv", index=False)

# Display transformed dataset
print(df_cleaned.head())
```

```python
data=pd.read_csv("transformed_transactions.csv")
```

```python
import seaborn as sns
corr_Data=data.corr()
sns.heatmap(corr_Data)
```

```python
from xgboost import XGBClassifier
# from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
X=data.drop(columns=["fraud"])
y=data["fraud"]
X
```

```python
X.columns
```

```python
x_train, x_test, y_train, y_test=train_test_split(X,y, test_size=0.3)
print(x_train.shape, x_test.shape, y_train.shape, y_test.shape)
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
x_train_scaled=scaler.fit_transform(x_train)
x_test_scaled=scaler.transform(x_test)
x_train_scaled
```

```python
from collections import Counter
from imblearn.over_sampling import SMOTE, ADASYN

# Check original class distribution
print("Original class distribution:", Counter(y_train))

#Oversampling (SMOTE)
smote = SMOTE(random_state=42)
x_train_resampled, y_train_resampled = smote.fit_resample(x_train, y_train)
print("After SMOTE (oversampling):", Counter(y_train_resampled))
```

```python
from xgboost import XGBClassifier
xgbc=XGBClassifier(random_state=42)
xgbc.fit(x_train_resampled, y_train_resampled)
xgbc_pred=xgbc.predict(x_test_scaled)
print("predictions: ", xgbc_pred)
```

```python
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
print("accuracy score: ", accuracy_score(y_test, xgbc_pred))
print("classification report: \n", classification_report(y_test, xgbc_pred, zero_division=1))
```

```python
x_train.columns
```

```python
# Hyperparameter tuning for Random Forest
from sklearn.model_selection import RandomizedSearchCV, StratifiedKFold


print("\nPerforming hyperparameter tuning...")
param_dist = {
    'n_estimators': [200, 300],
    'max_depth': [ 10, 20, 30],
    'min_samples_split': [2, 3, 5, 7],
    'min_samples_leaf': [1, 2, 3, 4],
    'max_features': ['sqrt', 'log2'],
    'criterion':['gini', 'entropy'],
}

cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)
# rf_tuned = RandomForestClassifier(random_state=42)
rf_tuned = XGBClassifier(random_state=42)

random_search = RandomizedSearchCV(
    estimator=rf_tuned,
    param_distributions=param_dist,
    scoring='accuracy',
    cv=cv,
    n_jobs=-1,
    n_iter=20,
)

random_search.fit(x_train_resampled, y_train_resampled)

print(f"Best parameters: {random_search.best_params_}")
print(f"Best F1 score: {random_search.best_score_:.4f}")

# Use the best model
best_rf_model = random_search.best_estimator_
y_pred_best = best_rf_model.predict(x_test_scaled)

print("\nTuned Random Forest Model Performance:")
print(classification_report(y_test, y_pred_best))
```
